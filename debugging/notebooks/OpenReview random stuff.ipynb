```python
import openreview
import os

pdf_dir = './iclr_pdfs/'

if not os.path.exists(pdf_dir):
    os.makedirs(pdf_dir)

def download_pdf(client, forum_id, pdf_dir):
    pdf_binary = client.get_pdf(forum_id)
    pdf_outfile = os.path.join(pdf_dir, '{}.pdf'.format(forum_id))
    with open(pdf_outfile, 'wb') as file_handle:
        file_handle.write(pdf_binary)

# guest_client = openreview.Client(baseurl='https://api.openreview.net')
# download_pdf(guest_client, "HylTBhA5tQ", pdf_dir)
```
