```python
import collections
import glob
import json

input_dir = "/Users/nnayak/Downloads/05-16-provisional-emnlp-release-unsplit"

CONFS = "ICLR2019 ICLR2020".split()

pre_split_data = {
    conf : collections.defaultdict(list) for conf in CONFS
}

for filename in glob.glob(input_dir + "/*"):
    mini_name = filename.split("/")[-1]
    with open(filename, 'r') as f:
        obj = json.load(f)
        
        pre_split_data[obj["metadata"]["conference"]][obj["metadata"]["forum_id"]].append(mini_name)
        
        
def is_single_annotated(annotations):
    annotation_counter = collections.Counter()
    for annotation in annotations:
        review_id, annotator, _ = annotation.split('.')
        annotation_counter[review_id] += 1
        
    is_single = set(annotation_counter.values()) == set([1])
    if is_single:
        return "def_train"
    elif set(annotation_counter.values()) == set([3]):
        return "def_test"
    else:
        #print(annotation_counter)
        return "?"
```

```python
splits = {
    conf : {"traindev": [], "test": []} for conf in CONFS}

allocation_map = {}

for conference, annotation_map in pre_split_data.items():
    qmark_forum_counter = 0
    qmark_review_counter = 0
    for forum_id, annotations in annotation_map.items():
        reviews = list(sorted(set(x.split(".")[0] for x in annotations)))
        if is_single_annotated(annotations) == "def_train":
            designated_set = "traindev"
        else:
            designated_set = "test"
            
        splits[conference][designated_set].append(reviews)
        for review in reviews:
            allocation_map[review] = (conference, designated_set)
            
```
