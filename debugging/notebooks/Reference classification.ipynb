```python
import collections
import re

def featurize(line):
    normalized_line = re.sub('[0-9]', '0', line)
    features = collections.Counter()
    for i in range(len(line) - 3):
        features[normalized_line[i:i+3]] += 1
    return features


featurize("Hello this is a 2019 example.")
```
