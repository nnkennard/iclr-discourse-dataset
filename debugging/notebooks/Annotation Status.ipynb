```python
COMPLETED, STARTED, NOT_STARTED = "completed started not_started".split()

def get_completed_rebuttals_from_assignments(assignments, example_map):
    review_statuses = collections.defaultdict(lambda: collections.defaultdict(list))
    rebuttal_statuses = collections.defaultdict(lambda: collections.defaultdict(list))
    for a in assignments:
        initials = a["fields"]["initials"]
        review_id = example_map[a["fields"]["example"]]
        if not a["fields"]["num_completed_sentences"]:
            rebuttal_statuses[review_id][NOT_STARTED].append(initials)
        elif a["fields"]["num_completed_sentences"] == a["fields"]["num_rebuttal_sentences"]:
            rebuttal_statuses[review_id][COMPLETED].append(initials)
        else:
            rebuttal_statuses[review_id][STARTED].append(initials)
        if a["fields"]["is_review_complete"]:
            print(review_id, initials)
            review_statuses[review_id][COMPLETED].append(initials)
        else:
            review_statuses[review_id][NOT_STARTED].append(initials)
            
    return review_statuses, rebuttal_statuses
```

```python
example_map = {}
for e in orda_text["example"]:
    example_map[e["pk"]] = e["fields"]["review_id"]

review_statuses, rebuttal_statuses = get_completed_rebuttals_from_assignments(orda_assignments["annotatorassignment"], example_map)
```

```python
# assignment_map = collections.defaultdict(list)

# for i in orda_assignments["annotatorassignment"]:
#     assignment_map[i["fields"]["review_id"]].append(i)

example_annotations = []

ExampleDetails = collections.namedtuple("ExampleDetails", 
        ("example_idx dataset interleaved_idx review_id rev_not_started rev_completed "
         "reb_not_started reb_started reb_completed total_completed").split())

for e in orda_text["example"]:
    review_id = e["fields"]["review_id"]
    example_annotations.append(
        ExampleDetails(e["pk"], e["fields"]["dataset"], e["fields"]["interleaved_index"], review_id,
                                review_statuses[review_id][NOT_STARTED],
                                review_statuses[review_id][COMPLETED],
                                rebuttal_statuses[review_id][NOT_STARTED],
                                rebuttal_statuses[review_id][STARTED],
                                rebuttal_statuses[review_id][COMPLETED],
                                len(rebuttal_statuses[review_id][COMPLETED]),
                                )._asdict()
    )
```

```python
b = pd.DataFrame.from_dict(example_annotations)
```
